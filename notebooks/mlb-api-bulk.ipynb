```python
import os
import requests
import pandas as pd
from datetime import datetime
import pytz

# === Time Setup ===
eastern = pytz.timezone("US/Eastern")
today = datetime.now(eastern).date()
date_str = today.isoformat()

# === Output paths (relative to notebook root) ===
base_dir = os.path.abspath(os.path.join("..", "data"))
pitching_dir = os.path.join(base_dir, "team_pitching")
batting_dir = os.path.join(base_dir, "team_batting")

os.makedirs(pitching_dir, exist_ok=True)
os.makedirs(batting_dir, exist_ok=True)

# === Get all MLB team IDs ===
def get_all_teams():
    url = "https://statsapi.mlb.com/api/v1/teams?sportId=1"
    res = requests.get(url)
    res.raise_for_status()
    return res.json().get("teams", [])

teams = get_all_teams()

# === Fetch team stats for all teams ===
def fetch_team_stats(team_id, group):
    url = f"https://statsapi.mlb.com/api/v1/teams/{team_id}/stats"
    params = {
        "stats": "season",
        "season": today.year,
        "group": group,
        "gameType": "R"
    }
    res = requests.get(url, params=params)
    res.raise_for_status()
    splits = res.json().get("stats", [])[0].get("splits", [])
    return splits[0]["stat"] if splits else {}

# === Collect and tag team stats ===
pitching_data = []
batting_data = []

for team in teams:
    team_id = team["id"]
    team_name = team["name"]

    print(f"📊 Fetching stats for {team_name}")

    try:
        pitching_stats = fetch_team_stats(team_id, "pitching")
        pitching_stats["team"] = team_name
        pitching_data.append(pitching_stats)
    except Exception as e:
        print(f"⚠️ Pitching error for {team_name}: {e}")

    try:
        batting_stats = fetch_team_stats(team_id, "hitting")
        batting_stats["team"] = team_name
        batting_data.append(batting_stats)
    except Exception as e:
        print(f"⚠️ Batting error for {team_name}: {e}")

# === Save DataFrames to CSV ===
pd.DataFrame(pitching_data).to_csv(os.path.join(pitching_dir, f"team_pitching_{date_str}.csv"), index=False)
pd.DataFrame(batting_data).to_csv(os.path.join(batting_dir, f"team_batting_{date_str}.csv"), index=False)

print("✅ All team stats saved successfully.")
```

```python
import os
import requests
import pandas as pd
from datetime import datetime
import pytz

# === Time Setup ===
eastern = pytz.timezone("US/Eastern")
today = datetime.now(eastern).date()
date_str = today.isoformat()

# === Output paths (relative to notebook root) ===
base_dir = os.path.abspath(os.path.join("..", "data"))
pitching_dir = os.path.join(base_dir, "team_pitching")
batting_dir = os.path.join(base_dir, "team_batting")

os.makedirs(pitching_dir, exist_ok=True)
os.makedirs(batting_dir, exist_ok=True)

# === Get all MLB team IDs ===
def get_all_teams():
    url = "https://statsapi.mlb.com/api/v1/teams?sportId=1"
    res = requests.get(url)
    res.raise_for_status()
    return res.json().get("teams", [])

teams = get_all_teams()

# === Fetch team stats only (no player data) ===
def fetch_team_stats(team_id, group):
    url = f"https://statsapi.mlb.com/api/v1/teams/{team_id}/stats"
    params = {
        "stats": "season",
        "season": today.year,
        "group": group,
        "gameType": "R"
    }
    res = requests.get(url, params=params)
    res.raise_for_status()
    splits = res.json().get("stats", [])[0].get("splits", [])
    return splits[0]["stat"] if splits else {}

# === Collect and tag team-level stats ===
pitching_data = []
batting_data = []

for team in teams:
    team_id = team["id"]
    team_name = team["name"]

    print(f"📊 Fetching team stats for {team_name}")

    try:
        pitching_stats = fetch_team_stats(team_id, "pitching")
        pitching_stats["team"] = team_name
        pitching_data.append(pitching_stats)
    except Exception as e:
        print(f"⚠️ Pitching error for {team_name}: {e}")

    try:
        batting_stats = fetch_team_stats(team_id, "hitting")
        batting_stats["team"] = team_name
        batting_data.append(batting_stats)
    except Exception as e:
        print(f"⚠️ Batting error for {team_name}: {e}")

# === Save DataFrames to CSV ===
pd.DataFrame(pitching_data).to_csv(os.path.join(pitching_dir, f"team_pitching_{date_str}.csv"), index=False)
pd.DataFrame(batting_data).to_csv(os.path.join(batting_dir, f"team_batting_{date_str}.csv"), index=False)

print("✅ Team-level pitching and batting stats saved successfully.")
```
